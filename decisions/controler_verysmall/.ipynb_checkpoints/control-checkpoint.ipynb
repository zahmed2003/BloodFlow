```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
import networkx as nx
import pandas as pd
```

```python
data_directory = '../../data/experiment_verysmall/'
data = np.load(data_directory + 'network_params/data_network.npy')
data = np.transpose(data, (1, 2, 0))
data.shape
```

```python
F = 250
start = 750

S = data[:,start:start+F,0]
D = data[:,start:start+F,1]

N = len(S)
b_max = (10**8)*np.ones(N)
b_init = np.zeros(N)
Q_max = 2000
```

```python
# plt.plot(S[1], label = 'supply')
# plt.plot(D[1], label = 'demand')
# plt.legend()
```

```python
Q = cp.Variable([N,N,F])
B = cp.Variable([N,F+1])
M = cp.Variable([N,F])

obj_func = cp.sum([cp.norm(10*cp.pos(m_k) + 1*cp.neg(m_k),2) for m_k in M])
obj = cp.Minimize(obj_func)

constr =  []

# IC
constr += [B[:,0] == np.zeros(N)]
        
# dynamics    
for t in range(F):
    for k in range(N):
        constr += [B[k,t+1] == B[k,t] + S[k,t] - D[k,t] + cp.sum(Q[k,:,t]) + M[k,t]]

# storage constraints
for t in range(F+1):
    constr +=[0<= B[:,t]]

# # ship limit
# constr +=[cp.norm(Q, 'inf')<=Q_max]

# no shipping
constr +=[Q == np.zeros([N,N,F])]

# # anti-sym of Q
# for k in range(N):
#     for l in range(N):
#         constr += [Q[k,l] == - Q[l,k]]

# # can't ship what you dont have
# for t in range(F):
#     for k in range(N):
#         constr +=[cp.sum(Q[:,k, t]) <= B[k, t+1]]
prob = cp.Problem(obj,constr)
prob.solve()
print(prob.value)
Q_opt = Q.value
B_opt = B[:,1:].value
M_opt = M.value
```

```python
# plt.plot(B_opt[0], label = 'b_o')
plt.plot(np.cumsum(S[0]-D[0]), label = 'true diff bank 0')
plt.plot(np.cumsum(S[1]-D[1]), label = 'true diff bank 1')
plt.plot(Q_opt[0,1], label = 'bank 1 to bank 0')
plt.legend()
```
