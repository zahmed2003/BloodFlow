```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
import networkx as nx
import pandas as pd
```

#### setup

```python
data_directory = '../data/experiment_small/'
data = np.load(data_directory + 'network_params/data_network.npy')
data = np.transpose(data, (1, 2, 0))
data.shape
```

```python
G = nx.read_graphml(data_directory + "network_params/graph.graphml")
W = nx.to_numpy_array(G,weight='weight')
W
```

```python
F = 1000
start = 0

S = data[:,start:start+F+1,0]
D = data[:,start:start+F+1,1]

N = len(S)
b_max = [2000,500,500,500]
b_init = np.zeros(N)
Q_max = 50
alpha = 1.0
gamma = 0.999

G = np.diag(gamma**np.arange(F))
```

```python
for n in range (N):
    plt.plot((S-D)[n], label = 'bank ' + str(n))
plt.legend()
```

#### creating the optimization

```python
B = cp.Variable([N,F+1])
Q = cp.Variable([N,N,F])
M = cp.Variable([N,F])

obj_func = cp.sum([cp.norm(G@m_k,1) for m_k in M]) + alpha*cp.sum([cp.norm(cp.multiply(W,Q[:,:,t]),2) for t in range(F)])
obj = cp.Minimize(obj_func)

constr =  []

# IC
constr += [B[:,0] == b_init]
# constr += [M>=0]


# max storage
for t in range(F+1):
    constr +=[B[:,t]<=b_max]
    
# dynamics    
for t in range(F):
    for k in range(N):
        constr += [B[k,t+1] == B[k,t] + S[k,t+1] - D[k,t+1] + cp.sum(Q[k,:,t])+ M[k,t]]

# storage constraints
for t in range(F+1):
    constr +=[0<= B[:,t]]


##### shipments
for k in range(N):
    for l in range(N):
        constr +=[Q[k,l] == -Q[l,k]]

for k in range(N):
    for t in range(F):
        constr += [cp.sum(Q[:,k,t]) <= B[k,t]]

constr += [cp.norm(Q,'inf') <= Q_max]

prob = cp.Problem(obj,constr)
prob.solve( solver = 'CLARABEL')
print(prob.value)
Q_opt = Q.value
B_opt = B[:,1:].value
M_opt = M.value
```

#### evaluating results

```python
time = np.arange(F)
plt.stackplot(time, B_opt, labels=[f"bank {i}" for i in range(N)], alpha=0.8)
plt.legend()
```

```python
for n in range(N):
    plt.plot(M_opt[n],label='Bank ' + str(n) + ' loan')
plt.legend()
plt.title('Blood Bank Loans')
plt.show()

for n in range(N):
    plt.plot(B_opt[n],label='Bank ' + str(n) + ' storage')
plt.legend()
plt.title('Blood Bank Storage')
plt.show()
```

```python
def get_total_error(M,W,Q,C,alpha,beta):
    f = len(M)
    error = 0
    
    for t in range(f):
        error += np.norm(M[t,:],1) + alpha*np.norm(np.multiply(W,Q[t,:,:]),2) + beta*np.norm(C[t,:],1)
        
    return error
```
